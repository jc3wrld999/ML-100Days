```python
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.optimizers import SGD

```

```python
# raw data loading
df = pd.read_csv('admission.csv')

# 데이터 전처리
# 결측치 처리
train_data = df.dropna(how='any')

# 이상치 처리

# 정규화
x_data = df.drop('admit', axis=1, inplace=False).values
t_data = df['admit'].values.reshape(-1, 1)

scaler_x = MinMaxScaler()
scaler_x.fit(x_data)

x_data_norm = scaler_x.transform(x_data)
print(x_data_norm)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.optimizers import SGD
# Flatten Layer를 이용해서 input layer를 표현 

keras_model = Sequential()

keras_model.add(Flatten(input_shape=(3, )))
keras_model.add(Dense(1, activation='sigmoid'))

# model 설정(model의 정확도 평가)
keras_model.compile(optimizer=SGD(learning_rate=1e-4), loss='binary_crossentropy', metrics=['accuracy'])

keras_model.fit(x_data_norm, t_data, epochs=1000, verbose=1)
```
