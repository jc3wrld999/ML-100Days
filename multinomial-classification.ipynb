```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.optimizers import SGD
from sklearn.preprocessing import MinMaxScaler
```

```python
# Raw Data Loading
df = pd.read_csv('bmi.csv', skiprows=3)
df.head()

# 데이터 전처리
# 1. 결측치 처리
# 2. 이상치 처리
# 3. 정규화
x_data = df[['height', 'weight']].values # 2차원 형태의 array(독립변수 2개)
t_data = df['label'].values.reshape(-1, 1)

print(x_data)
scaler = MinMaxScaler()
scaler.fit(x_data)
x_data_norm = scaler.transform(x_data)
print(x_data_norm)

# one-hot encoding 처리를 해야함
# 종속변수 0 => 1 0 0
# 종속변수 1 => 0 1 0
# 종속변수 2 => 0 0 1
# keras로 처리
```

```python
# 모델을 구현
model = Sequential()

model.add(Flatten(input_shape=(2,)))
model.add(Dense(3, activation='softmax'))

model.compile(optimizer=SGD(learning_rate=1e-2), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(x_data_norm, t_data, epochs=100, batch_size=100, verbose=1, validation_split=0.2)
```
