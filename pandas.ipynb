### 데이터 셋 생성, 수정, 삭제
```python
DataFrame.drop(labels=None, axis=0, index=None, colums=None, level=None, inplace=False, errors='raise')
```
- axis: DataFrame의 row를 삭제할 때는 axis=0, column을 삭제할 때는 axis=1로 설정
- inplace: 원본 DataFrame은 유지하고 드롭된 DataFrame을 객체 변수로 받고 싶다면 False로 설정(default = false)
- labels: axis=1(columns삭제)일때 컬럼 이름 / axis=1(rows삭제)일때 인덱스 번호

## 인덱싱, 필터링


````python
# 인덱싱
# []
display(titanic_df['Name'])
display(titanic_df[['Name', 'Age']])
# loc[], iloc[] 
display(titanic_df.iloc[0,0]) # 1번째 행 1번째 열
display(titanic_df.loc[0,'PassengerId'])
# boolean indexing
display(titanic_df[titanic_df['Age']>60])
````

````python
data = {'Name': ['Chulmin', 'Enkyung', 'Jinwoong', 'Soobeom'],
       'Year': [2011, 2016, 2015, 2015],
       'Gender': ['Male', 'Female', 'Male', 'Male']
       }
data_df = pd.DataFrame(data, index=['one', 'two', 'three', 'four'])
data_df
````

````python
data_df.iloc[0, 0]
````

````python
data_df.iloc[0, 'Name'] # error
data_df.iloc['one', 0] # error
data_df.iloc
````

````python
print(data_df.iloc[1, 0]) # 두번째 행의 첫번째 열값
print(data_df.iloc[2, 1]) # 세번째 행의 두번째 열값
print(data_df.iloc[0:2, [0, 1]]) # 1~2 행의 첫번째, 두번째 열값
print(data_df.iloc[0:2, 0:3]) # 1~2행의 1~3번째 값
print(data_df.iloc[:, -1]) # 맨 마지막 컬럼 데이터
print(data_df.iloc[:, :-1]) # 맨 마지막 컬럼을 제외한 모든 데이터
print(data_df.iloc[:]) # 모든 데이터
print(data_df.iloc[:, :]) #모든 데이터
````
